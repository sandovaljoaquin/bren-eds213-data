### Using a cursor 

```python
# Create a cursor 
cur = conn.cursor()
```

```python
cur.execute("SELECT * FROM Site")
```

```python
# You can apply the fetchall().. 
# Idea of cursor is to go row by row 
cur.fetchall()
```

```python
# Getting one row 
cur.execute("SELECT * FROM Site")
cur.fetchone()
```

```python
cur.fetchone()[0]
```

```python
# We want to iterate throughout rows and print the site code
cur.execute("SELECT * FROM Site")

# 
for row in cur.fetchall(): 
    print(row[0])
```

```python
cur.execute("SELECT * FROM Site")
while True: 
    row = cur.fetchone()
    if row == None: 
        break 
    print(f"Site code: {row[0]}")
```

.fetchall() is very slow vs using a cursor - more efficient (one at a time or a few at a time)

```python
# With cursor approach, you can create tables 

cur.execute("""CREATE TEMP TABLE temp_nests AS 
            SELECT * FROM Bird_nests""")
```

```python
cur.execute("SELECT * FROM temp_nests LIMIT 5").fetchall()
```
